```python
import numpy as np
import matplotlib.pyplot as plt

# Sentence and settings
sentence = ["I", "like", "pizza"]
positions = list(range(len(sentence)))
d_model = 4  # embedding dimension
```

```python
# Positional encoding function
def positional_encoding(pos, d_model):
    pe = np.zeros(d_model)
    for i in range(d_model):
        div_term = 10000 ** (2 * (i // 2) / d_model)
        if i % 2 == 0:
            pe[i] = np.sin(pos / div_term)
        else:
            pe[i] = np.cos(pos / div_term)
    return pe
```

```python
# Generate positional encodings
positional_encodings = np.array([positional_encoding(pos, d_model) for pos in positions])

# Print positional encoding vectors
for word, encoding in zip(sentence, positional_encodings):
    print(f"{word}: {encoding}")
```

```python
# Plotting the positional encoding vectors
plt.figure(figsize=(10, 5))
for i, encoding in enumerate(positional_encodings):
    plt.plot(range(d_model), encoding, marker='o', label=f"{sentence[i]} (pos {i})")

plt.title("Positional Encoding Vectors")
plt.xlabel("Embedding Dimension Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

# Home Task 18

- Simulates 10,000 student scores from the given bivariate normal distribution.

- Plots a scatter plot of Math vs Physics scores.

- Calculates and compares the sample correlation with the theoretical one (œÅ = 0.7).

- Calculates the proportion of students scoring above 80 in both subjects.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
mu = [70, 75]  # Means for Math and Physics
sigma_math = 10
sigma_physics = 12
rho = 0.7

# Covariance matrix
cov = [
    [sigma_math**2, rho * sigma_math * sigma_physics],
    [rho * sigma_math * sigma_physics, sigma_physics**2]
]

# Simulate 10,000 samples
samples = np.random.multivariate_normal(mu, cov, 10000)
math_scores = samples[:, 0]
physics_scores = samples[:, 1]

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(math_scores, physics_scores, alpha=0.5, s=10, color='blue')
plt.xlabel('Math Scores')
plt.ylabel('Physics Scores')
plt.title('Scatter Plot of Simulated Math vs Physics Scores')
plt.grid(True)
plt.show()

# Sample correlation
sample_correlation = np.corrcoef(math_scores, physics_scores)[0, 1]
print(f"Sample correlation coefficient: {sample_correlation:.4f}")
print(f"Theoretical correlation coefficient: {rho}")

# Proportion scoring above 80 in both
count_above_80 = np.sum((math_scores > 80) & (physics_scores > 80))
proportion_above_80 = count_above_80 / 10000
print(f"Proportion of students scoring above 80 in both subjects: {proportion_above_80:.4f}")
```

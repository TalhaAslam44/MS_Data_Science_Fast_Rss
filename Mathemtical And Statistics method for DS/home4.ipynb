# importing liberaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

- Data matrix

```python

# Original data as a NumPy array (5 rows, 4 columns)
data = np.array([
    [25, 40000, 50, 10000],
    [30, 50000, 60, 15000],
    [22, 35000, 45, 8000 ],
    [28, 45000, 55, 12000],
    [26, 42000, 53, 11000]
], dtype=float)
```

```python
data
```

- Applying min-max scalar

```python

# 1) Apply MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

```

```python
scaler
```

```python
scaled_data
```

- Perfmoring PCA

```python

# 2) Perform PCA (reduce to 2D)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data
```

## ploting 2D Scatter

```python

# 3) Plot 2D scatter
plt.scatter(pca_data[:, 0], pca_data[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Customers (2D)')
plt.show()
```
